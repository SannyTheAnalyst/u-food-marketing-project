## Business Objective

The objective of this analysis is to explore customer demographics and purchasing behavior in a food marketing dataset to identify patterns that can support customer segmentation and improve marketing campaign effectiveness.


```python
import pandas as pd

df = pd.read_csv("../data/u_food_marketing.csv")
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum().sort_values(ascending=False).head(10)
```

```python
df["Income"].dropna().hist(bins=30)
```

```python
import matplotlib.pyplot as plt

plt.figure()
df["Income"].dropna().hist(bins=30)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Number of Customers")
plt.show()
```

```python
import matplotlib.pyplot as plt

category_totals = {
    "Wines": df["MntWines"].sum(),
    "Fruits": df["MntFruits"].sum(),
    "Meat": df["MntMeatProducts"].sum(),
    "Fish": df["MntFishProducts"].sum(),
    "Sweets": df["MntSweetProducts"].sum(),
    "Gold": df["MntGoldProds"].sum()
}

s = pd.Series(category_totals).sort_values()
plt.figure()
s.plot(kind="barh")
plt.title("Total Spend by Product Category")
plt.xlabel("Total Spend")
plt.ylabel("Category")
plt.show()
```

```python
accept_rate = df["AcceptedCmpOverall"].mean()
accept_rate
```

```python
plt.figure()
df["AcceptedCmpOverall"].value_counts().plot(kind="bar")
plt.title("Campaign Acceptance (0 = No, 1 = Yes)")
plt.xlabel("AcceptedCmpOverall")
plt.ylabel("Number of Customers")
plt.show()
```
